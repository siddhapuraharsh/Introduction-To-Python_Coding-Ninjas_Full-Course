```python
'''
Check Armstrong

Write a Program to determine if the given number is Armstrong number or not. Print true if number is armstrong, otherwise print false.
An Armstrong number is a number (with digits n) such that the sum of its digits raised to nth power is equal to the number itself.

For example,
371, as 3^3 + 7^3 + 1^3 = 371
1634, as 1^4 + 6^4 + 3^4 + 4^4 = 1634

Input Format :
Integer n
Output Format :
true or false

Sample Input 1 :
1
Sample Output 1 :
true

Sample Input 2 :
103
Sample Output 2 :
false
'''

# Solution-1

import math

def isArmstrong(num):
    
    temp_num = num
    armNo = 0
    length = len(str(num))
    
    while(num):
        armNo = armNo + math.pow(num%10, length)
        num = num//10
        
    if(temp_num == armNo):
        print('true')
    else:
        print('false')

n = int(input())
isArmstrong(n)



#Solution-2

def checkArmstrong(n):
    
  n_list = list(n)
  length = len(n_list)
  sum = 0
    
  for digit in n_list:
    sum += int(digit)**length
        
  if sum == int(n):
    return 'true'
  else:
    return 'false'
        
n = input()
print(checkArmstrong(n))
```

```python
'''
Zeros and Stars Pattern

Print the following pattern

Pattern for N = 4
*000*000*
0*00*00*0
00*0*0*00
000***000

Input Format :
N (Total no. of rows)
Output Format :
Pattern in N lines

Sample Input 1 :
3
Sample Output 1 :
*00*00*
0*0*0*0
00***00

Sample Input 2 :
5
Sample Output 2 :
*0000*0000*
0*000*000*0
00*00*00*00
000*0*0*000
0000***0000
'''


# Solution-1

def zeroStarPattern(n):
  answerMatrix = [ [0]*(2*n+1) for i in range(n) ]
    
  for i in range(n):
    for j in range(n):
      if i==j:
        answerMatrix[i][j] = '*'
                
  for i in range(n):
    answerMatrix[i][n] = '*'
        
  for i in range(n):
    for j in range(n+1, (2*n+1), 1):
      if i+j == 2*n:
        answerMatrix[i][j] = '*'
                
  for i in answerMatrix:
    for j in i:
      print(j, end='')
    print()

n = int(input())
zeroStarPattern(n)



# Solution-2

n = int(input())
rows = 0
while(rows <= n-1):
    
    zero = 1
    while(zero <= rows):
        print('0', end='')
        zero += 1
    
    print('*', end='')
    
    zero = 1
    while(zero <= n-rows-1):
        print('0', end='')
        zero += 1
   
    print('*', end='')

    
    zero = 1
    while(zero <= n-rows-1):
        print('0', end='')
        zero += 1
        
    print('*', end='')
    
    zero = 1
    while(zero <= rows):
        print('0', end='')
        zero += 1
    
    print()
    rows += 1
```

```python
'''
Number Star pattern 1
Print the following pattern for given number of rows.

Input format :
Integer N (Total number of rows)
Output Format :
Pattern in N lines

Sample Input :
   5
Sample Output :
 5432*
 543*1
 54*21
 5*321
 *4321
'''


# Solution-1

def numStarPattern(n):
  for i in range(1, n+1):
    for j in range(n, 0, -1):
      if i==j:
        print('*', end='')
      else:
        print(j, end='')
    print()

n = int(input())
numStarPattern(n)



# Solution-2

n = int(input())

for i in range(1, n+1):
  for j in range(0, n-i):
    print(n-j, end='')
  print('*', end='')
  for j in range(n-i+1, n, 1):
    print(n-j, end='')
  print()
```
